WeRateDogs Twitter account has over 9 million followers and has received international media coverage. Due to the popularity of WeRateDogs, we decided to look into the statistics of the account and extract some data from the tweets. Omitting information about how the data was gathered and cleaned, one major assumption made in the cleaning the data derived from WeRateDogs is we assumed that the maximum value of the rating numerator is 15. Therefore the outliers in the rating numerator column was changed to 15. With these changes made to the rating numerator column, we created a standard rating column by dividing the rating numerator by the rating denominator. Another important thing to note is that while performing exploartory data analysis on the data derived from WeRateDogs, we removed outliers in the retweet count column and the favourite count column. Without futher delay, let us dive into the visualiztion made from the data derived from WeRateDogs.

 

From the histogram above, we can see that the standard rating column is skewed to the left. This means the standard rating column has a mean (1.06) which is less than it's median (1.10), meaning that more dogs received a high rating.

 

Contrary to the standard rating column, both retweet and favourite column are skewed to the right, meaning that the mean of both columns, 4387.2 and 1519.7 respectively are higher than the median of the columns, 3172.0 and 1128.0 respecvtively. The implication of this (considering the fact that the standard rating column is skewed to the left) is that even though more dogs receive a high rating, only some of the dogs who receive a high rating has a high number of retweets and favourite counts.

 

It is interesting to know that both standard rating with retweet count has weak correlation of 0.3, and standard rating with favourite count also has a weak correlation of 0.37. The implication of this(taking into consideration the skewness of the retweets, favourite and standard rating histogram) is that high rating does not necessarily result in high number of retweet and favourite count, but few high rating tend to have high number of retweet and favourite counts
